# Импорты

```python
import pandas as pd
import numpy as np
import plotly as pt
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import json
```

```python
age = pd.DataFrame({'Мужчины':[0, 2, 50, 36, 9, 2, 1], 'Женщины':[0, 2, 61, 29, 7, 1, 0]},
                   index=['До 20', '20-25', '26-35', '36-45', '46-55', '56-60', '>60'])

display(age)

experience = pd.DataFrame({'Мужчины':[3, 9, 23, 18, 19, 13, 15],
                           'Женщины':[3, 13, 29, 20, 17, 9, 9]},
                          index=['До 1 года', '1-3', '3-6', '6-9', '9-12', '12-15', '> 15 лет'])

display(experience)

education = pd.DataFrame({'Образование': [1, 5, 9, 85]},
                         index=['Среднее', 'Среднее специальное', 'Незаконченное высшее', 'Высшее'])

display(education)

gender = pd.DataFrame({'Пол': [80, 20]}, index=['Мужчины', 'Женщины'])
display(gender)
```

# Графики

```python
#fig = make_subplots(rows=1, cols=3, subplot_titles=("<b>Возраст</b>", "<b>Опыт работы</b>", "<b>Образование</b>"), specs=[[{'type': 'histogram'}, {'type': 'histogram'}, {'type': 'pie'}]])

labels1 = age.index
values11 = age['Женщины']
values12 = age['Мужчины']

#fig = go.Figure()
fig.add_trace(go.Bar(y = labels1, x = values11, orientation='h', text=values11, textposition='outside', name='Женщины'),  row=1, col=1)
fig.add_trace(go.Bar(y = labels1, x = values12, orientation='h', text=values12, textposition='outside', name='Мужчины'),  row=1, col=1)
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  #title="<b>Возраст</b>",
                  yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))


labels2=experience.index
values21 = experience['Женщины']
values22 = experience['Мужчины']

#fig = go.Figure()
fig.add_trace(go.Bar(y = labels2, x = values21, orientation='h', text=values21, textposition='outside', name='Женщины'),  row=1, col=2)
fig.add_trace(go.Bar(y = labels2, x = values22, orientation='h', text=values22, textposition='outside', name='Мужчины'),  row=1, col=2)
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  #title="<b>Опыт работы</b>",
                  yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))

labels3=education.index
values3 = education['Образование']

#fig = go.Figure()
#pull = [0]*len(values)
#pull[values.tolist().index(values.max())] = 0.2
fig.add_trace(go.Pie(values=values3, labels=labels3,  hole=0.8), row=1, col=3)  #pull=pull,
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  #title="<b>Образование</b>",
                  #yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))

fig.show()
```

```python
labels1 = age.index
values11 = age['Женщины']
values12 = age['Мужчины']

fig = go.Figure()
fig.add_trace(go.Bar(y = labels1, x = values11, orientation='h', text=values11, textposition='outside', name='Женщины'))
fig.add_trace(go.Bar(y = labels1, x = values12, orientation='h', text=values12, textposition='outside', name='Мужчины'))
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  title="<b>Возраст</b>",
                  yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))
fig.show()

labels2=experience.index
values21 = experience['Женщины']
values22 = experience['Мужчины']

fig = go.Figure()
fig.add_trace(go.Bar(y = labels2, x = values21, orientation='h', text=values21, textposition='outside', name='Женщины'))
fig.add_trace(go.Bar(y = labels2, x = values22, orientation='h', text=values22, textposition='outside', name='Мужчины'))
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  title="<b>Опыт работы</b>",
                  yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))
fig.show()

labels3=education.index
values3 = education['Образование']

fig = go.Figure()
#pull = [0]*len(values)
#pull[values.tolist().index(values.max())] = 0.2
fig.add_trace(go.Pie(values=values3, labels=labels3,  hole=0.8))  #pull=pull,
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  title="<b>Образование</b>",
                  #yaxis = dict(autorange="reversed"),
                  margin=dict(l=0, r=0, t=50, b=0))

fig.show()

labels4=gender.sort_values(by='Пол').index
values4 = gender['Пол']

fig4 = go.Figure()
fig4.add_trace(go.Bar(x = [values4['Женщины']], orientation='h', text=str(values4['Женщины'])+'%', textposition='outside', name='Женщины')) #y = labels[0],
fig4.add_trace(go.Bar(x = [values4['Мужчины']], orientation='h', text=str(values4['Мужчины'])+'%', textposition='outside', name='Мужчины')) #y = labels[1],

fig4.update_layout(#legend_orientation="h",
                   #legend=dict(x=.5, xanchor="center"),
                   barmode='stack',
                   title="<b>Пол</b>",
                   height=90,
                   yaxis_visible=False,
                   xaxis_visible=False,
                   margin=dict(l=0, r=0, t=50, b=0)
                  )
fig4.show()
```
